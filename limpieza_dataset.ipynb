## Limpieza dataset Handwashing with soap

https://apps.who.int/gho/data/node.main.WSHHYGIENE?lang=en

The percentage of population living in households that have a handwashing facility with soap and water at home. Handwashing facilities can consist of a sink with tap water, but can also include other devices that contain, transport or regulate the flow of water. Buckets with taps, tippy-taps and portable basins are all examples of handwashing facilities. Bar soap, liquid soap, powder detergent and soapy water all count as soap for monitoring purposes.

```python
water = pd.read_csv('water.csv')
water.head()
```

```python
water.columns = ['countriesAndTerritories', 'percentage_handwashing_facility_2020']
```

```python
water.head()
```

```python
water.drop([0,1],axis=0)
```

```python
water.isna().sum()
```

```python
countries2=pd.DataFrame(water.groupby(["countriesAndTerritories"], sort=True)["countriesAndTerritories"].count()
                              .reset_index(name="total"))
countries2 
```

### Merge de ambos datasets

```python
merge = data1.merge(water, on=["countriesAndTerritories"], how="left")
merge.head()
```

```python
merge.tail()
```

```python
merge.isna().sum()
```

```python
merge.info()
```

### Elimino nulos

```python
merge = merge.dropna()
merge.head()
```

```python
merge.isna().sum()
```

```python
merge.shape
```

```python
countries3=pd.DataFrame(merge.groupby(["countriesAndTerritories"], sort=True)["countriesAndTerritories"].count()
                              .reset_index(name="total"))
countries3
```

```python
cont=pd.DataFrame(merge.groupby(["continentExp"], sort=True)["continentExp"].count()
                              .reset_index(name="total"))
cont
```

```python
europe = merge[(merge.continentExp == 'Europe')]
europe
```

```python
eurcountries=pd.DataFrame(europe.groupby(["countriesAndTerritories"], sort=True)["countriesAndTerritories"].count()
                              .reset_index(name="total"))
eurcountries
```
